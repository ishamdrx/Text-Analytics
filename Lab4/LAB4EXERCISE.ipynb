```python
# Load dataset
import pandas as pd
file_path = "UNITENReview.csv"
df = pd.read_csv(file_path)
# Display column content without truncation
pd.set_option('display.max_colwidth', None) # Set to None for unlimited width
print(df)
```

```python
# Lowercase conversion
def convert_to_lowercase(text):
 return text.lower()
df["lowercased"] = df["Review"].apply(convert_to_lowercase)
# Display column content without truncation
pd.set_option('display.max_colwidth', None) # Set to None for unlimited width
print(df["lowercased"])
```

```python
import re

# Dictionary of slang words and their replacements
slang_dict = {
    "tbh": "to be honest",
    "omg": "oh my god",
    "lol": "laugh out loud",
    "idk": "i don't know",
    "brb": "be right back",
    "btw": "by the way",
    "imo": "in my opinion",
    "smh": "shaking my head",
    "fyi": "for your information",
    "np": "no problem",
    "ikr": "i know right",
    "asap": "as soon as possible",
    "bff": "best friend forever",
    "gg": "good game",
    "hmu": "hit me up",
    "rofl": "rolling on the floor laughing"
}

# Function to replace slang words
def replace_slang(text):
    words = text.split()
    replaced_text = " ".join([slang_dict.get(word, word) for word in words])
    return replaced_text

# Apply the function
df["slangs_replaced"] = df["lowercased"].apply(replace_slang)

# Display results
print(df[["lowercased", "slangs_replaced"]].head())
```

```python
from nltk.tokenize import word_tokenize

# Dictionary of common contractions
contractions_dict = {
    "wasn't": "was not",
    "isn't": "is not",
    "aren't": "are not",
    "weren't": "were not",
    "doesn't": "does not",
    "don't": "do not",
    "didn't": "did not",
    "can't": "cannot",
    "couldn't": "could not",
    "shouldn't": "should not",
    "wouldn't": "would not",
    "won't": "will not",
    "haven't": "have not",
    "hasn't": "has not",
    "hadn't": "had not",
    "i'm": "i am",
    "you're": "you are",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "we're": "we are",
    "they're": "they are",
    "i've": "i have",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "i'd": "i would",
    "you'd": "you would",
    "he'd": "he would",
    "she'd": "she would",
    "we'd": "we would",
    "they'd": "they would",
    "i'll": "i will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "we'll": "we will",
    "they'll": "they will",
    "let's": "let us",
    "that's": "that is",
    "who's": "who is",
    "what's": "what is",
    "where's": "where is",
    "when's": "when is",
    "why's": "why is"
}

# Function to replace contractions
def replace_contractions(text):
    words = word_tokenize(text)
    replaced_text = " ".join([contractions_dict.get(word.lower(), word) for word in words])
    return replaced_text

# Apply the function
df["contractions_replaced"] = df["slangs_replaced"].apply(replace_contractions)

# Display results
print(df[["slangs_replaced", "contractions_replaced"]].head())
```

```python
import string

# Function to remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Apply the function
df["punctuations_removed"] = df["contractions_replaced"].apply(remove_punctuation)

# Display results
print(df[["contractions_replaced", "punctuations_removed"]].head())
```

```python
# Function to remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

# Apply the function
df["numbers_removed"] = df["punctuations_removed"].apply(remove_numbers)

# Display results
print(df[["punctuations_removed", "numbers_removed"]].head())
```

```python
from autocorrect import Speller

# Initialize Speller
spell = Speller(lang='en')

# Function to correct spelling
def correct_spelling(text):
    return " ".join([spell(word) for word in text.split()])

# Apply the function
df["spelling_corrected"] = df["numbers_removed"].apply(correct_spelling)

# Display results
print(df[["numbers_removed", "spelling_corrected"]].head())
```

```python
from nltk.corpus import stopwords

# Define stopwords list
stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(text):
    words = text.split()
    return " ".join([word for word in words if word.lower() not in stop_words])

# Apply the function
df["stopwords_removed"] = df["spelling_corrected"].apply(remove_stopwords)

# Display results
print(df[["spelling_corrected", "stopwords_removed"]].head())
```

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# Function to apply stemming
def stem_text(text):
    words = text.split()
    return " ".join([stemmer.stem(word) for word in words])

# Apply the function
df["stemmed_words"] = df["stopwords_removed"].apply(stem_text)

# Display results
print(df[["stopwords_removed", "stemmed_words"]].head())
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to map POS tags
def get_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Function to lemmatize text
def lemmatize_text(text):
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    return " ".join([lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags])

# Apply the function
df["lemmatized"] = df["stopwords_removed"].apply(lemmatize_text)

# Display results
print(df[["stopwords_removed", "lemmatized"]].head())
```

```python
df.to_csv("Processed_UNITENReview.csv", index=False, encoding="utf-8")
print("Pre-processing complete. File saved as Processed_UNITENReview.csv")
```
